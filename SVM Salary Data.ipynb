# Data Preparation

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
salary_data_test['Salary']   =le.fit_transform(salary_data_test['Salary'])
salary_data_train['Salary']  =le.fit_transform(salary_data_train['Salary'])
```

```python
salary_data_train_1 = pd.get_dummies(salary_data_train)
salary_data_test_1  = pd.get_dummies(salary_data_test)
salary_data_train_1
```

```python
#Min-Max Scalar
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
salary_data_train_1 = norm_func(salary_data_train_1.iloc[:,0:])
salary_data_test_1 = norm_func(salary_data_test_1.iloc[:,0:])
salary_data_train_1
```

```python
x_train = salary_data_train_1.drop('Salary',axis=1)
y_train = salary_data_train_1[['Salary']]
x_test  = salary_data_test_1.drop('Salary',axis=1)
y_test  = salary_data_test_1[['Salary']] 
```

# Model Building

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import SVC
s = SVC(kernel='rbf')
classifier = s.fit(x_train,y_train)
```

```python
# Model Testing
y_pred = classifier.predict(x_test)
```

```python
# Model Evalution
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
print('Accuracy Score    : ', accuracy_score(y_test,y_pred))
print('precision Score   : ', precision_score(y_test,y_pred))
print('Recall Score      : ',recall_score(y_test,y_pred))
print('Confussion matrix : \n',confusion_matrix(y_test,y_pred))
```
